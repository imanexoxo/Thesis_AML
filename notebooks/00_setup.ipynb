## 8. Configurazione
### Random seed
Il random seed è la base dei generatori dei numeri casuali: usando sempre lo stesso seed, si otterranno sempre gli stessi numeri casuali che rendono il codice riproducibile

```python
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED) # imposto il random seed per NumPy
torch.manual_seed(RANDOM_SEED) # imposto il random seed per PyTorch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
ROOT = Path('..')

print(f"Seed: {RANDOM_SEED}")
print(f"Device: {device}")
print(f"Project root: {ROOT}")
print("\nFine setup iniziale")
```
